# Análisis Exploratorio de los datos
Los estadísticos descriptivos más habitualmente utilizados han sido la media y la desviación típica. Sin embargo, el uso automático de estos índices no es muy aconsejable. Se hallan los valores máximos, media, mediana (percentil 50%), moda, desviación típica, varianza, percentil inferior 25% y percentil superior 75%

```python
total=pd.read_excel("TOTAL.xlsx",index_col=0)
total
```

```python
maximos = total.max() #Serie que contiene los valores máximos de cada columna con función max()
maximos
```

```python
media=total.mean() # media aritmetica de cada columna con función función mean()   “No se por qué aparece así  ?????”
media
```

```python
mediana=total.median() # mediana de cada columna utilizando la función median() 
mediana
```

```python
stats.mode(total) # Moda de cada columna # el 2do array devuelve la frecuencia. NO EJECUTAR
```

```python
desvtipica=np.std(total, 0) # Desviación típica de cada columna utilizando la función std ()
desvtipica
```

```python
varianza=np.var(total, 0) # varianza de cada columna utilizando la función var()
varianza
```

```python
resumenestad=total.describe() # Resumen estadistadistico con pandas 
resumenestad #cantidad,media,desviación, mímino, percentil inferior, 50% o mediana, percentil superior, máximo
```


Muchas veces los indicadores de la estadística descriptiva no nos proporcionan una imagen clara de nuestros datos. Por esta razón, siempre es útil complementarlos con gráficos de las distribuciones de los datos, que describan con qué frecuencia aparece cada valor. La representación más común de una distribución es un histograma, que es un gráfico que muestra la frecuencia o probabilidad de cada valor. 

```python
a=plt.hist(total.VENTAS,color="red") #Histograma de una variable utilizando la función hist()
a=plt.xlabel("Valores")
a=plt.ylabel("Frecuencia");
a=plt.title("Histograma VENTAS")
a

plt.savefig("HistogramaVENTAS.png") #Guarda la imagen como png
```

```python
a = sns.pairplot(total, vars=["VAL_AGREGADO","VENTAS","EXPORTACIONES"],kind="reg") 
a # Resumen gráfico con histogramas y diagramas de dispersión de las variables SOCIOSMUJ y SOCIOSHOM
a.savefig("RESULTADOSFINANCIEROS.png") #Guarda la imagen como png
```
